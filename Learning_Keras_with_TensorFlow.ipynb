```python
# installing libraries
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

# Making Data

```python
train_labels = []
train_samples = []
```

```python
for i in range(50):
    # 5% young people who experienced side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)

    # 5% old people who experienced with no side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)

for i in range(1000):
    # 95% young people who did not experience side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)

    # 95% old people who experienced side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

# Building Model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
physical_devices = tf.config.list_logical_devices('GPU')
print('Num GPUs Available: ', len(physical_devices))
```

```python
import tensorflow as tf
if tf.test.gpu_device_name():
    print('Default GPU Device: {}'.format(tf.test.gpu_device_name()))
else:
    print("Please install GPU version of TF")
```

```python
# the input_shape in 1st layer tells Keras what type of input to expect
model = Sequential([
                    Dense(units=16, input_shape=(1,), activation='relu'),
                    Dense(units=32, activation='relu'),
                    Dense(units=2, activation='softmax')
])
# the units=2 in the last layer (output layer) tells there's 2 output classes for classification
```

```python
model.summary()
```

# Training our Model

```python
model.compile(
    optimizer = Adam(learning_rate=0.0001),
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(
    x = scaled_train_samples,
    y = train_labels,
    batch_size = 10,
    epochs = 30,
    shuffle = True,
    verbose = 2
)
```

# Creating a Validation Set

```python
model.fit(
    x = scaled_train_samples,
    y = train_labels,
    validation_split = 0.1, # to keep a check on overfitting
    batch_size = 10,
    epochs = 30,
    shuffle = True, # even more important when using validation set
    verbose = 2
)
```

# Predicting with our Model

## Making Test Set

```python
test_labels = []
test_samples = []
```

```python
for i in range(10):
    # 5% young people who experienced side effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)

    # 5% old people who experienced with no side effects
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(0)

for i in range(200):
    # 95% young people who did not experience side effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)

    # 95% old people who experienced side effects
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(1)
```

```python
test_labels = np.array(test_labels)
test_samples = np.array(test_samples)
test_labels, test_samples = shuffle(test_labels, test_samples)
```

```python
scaled_test_samples = scaler.fit_transform(test_samples.reshape(-1,1))
```

## Predicting with our Model

```python
predictions = model.predict(
    x = scaled_test_samples,
    batch_size = 10,
    verbose = 0
)
```

```python
for i in predictions:
    print(i)
```

Each sample gets 2 probabilities for experiencing a side effect or not.

```python
rounded_predictions = np.argmax(predictions, axis=-1)
```

```python
for i in rounded_predictions:
    print(i)
```

# Creating a Confusion Matrix

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(
   y_true=test_labels, y_pred=rounded_predictions)

plt.show()
```
